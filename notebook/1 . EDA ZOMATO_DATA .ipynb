### 2.2 Dataset information

- ID: Unique identifier for each entry.
- Delivery_person_ID: Unique identifier for delivery persons.
- Delivery_person_Age: Age of the delivery person (in years).
- Delivery_person_Ratings: Ratings of the delivery person.
- Restaurant_latitude: Latitude of the restaurant.
- Restaurant_longitude: Longitude of the restaurant.
- Delivery_location_latitude: Latitude of the delivery location.
- Delivery_location_longitude: Longitude of the delivery location.
- Order_Date: Date of the order.
- Time_Orderd: Time when the order was placed (in datetime format).
- Time_Order_picked: Time when the order was picked up.
- Weather_conditions: Weather conditions at the time of order.
- Road_traffic_density: Traffic density at the time of order.
- Vehicle_condition: Condition of the vehicle.
- Type_of_order: Type of order (e.g., online, phone).
- Type_of_vehicle: Type of vehicle used for delivery.
- multiple_deliveries: Indicator of whether multiple deliveries were made.
- Festival: Indicator of whether it was a festival day.
- City: City where the order was placed.
- Time_taken (min): Time taken for delivery (in minutes).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set(style="whitegrid")

# Univariate Analysis for numerical variables
# Plot histograms or density plots
plt.figure(figsize=(18, 6))

# Histogram for Delivery_person_Age
plt.subplot(1, 3, 1)
sns.histplot(df['Delivery_person_Age'], kde=True, color='skyblue')
plt.title('Distribution of Delivery Person Age')

# Histogram for Delivery_person_Ratings
plt.subplot(1, 3, 2)
sns.histplot(df['Delivery_person_Ratings'], kde=True, color='salmon')
plt.title('Distribution of Delivery Person Ratings')

# Histogram for Time_taken (min)
plt.subplot(1, 3, 3)
sns.histplot(df['Time_taken (min)'], kde=True, color='green')
plt.title('Distribution of Time Taken for Delivery (min)')

plt.tight_layout()
plt.show()

# Univariate Analysis for categorical variables
# Plot bar plots
plt.figure(figsize=(18, 12))

# Bar plot for Weather_conditions
plt.subplot(2, 3, 1)
sns.countplot(data=df, x='Weather_conditions', palette='pastel')
plt.title('Distribution of Weather Conditions')

# Bar plot for Road_traffic_density
plt.subplot(2, 3, 2)
sns.countplot(data=df, x='Road_traffic_density', palette='pastel')
plt.title('Distribution of Road Traffic Density')

# Bar plot for Type_of_order
plt.subplot(2, 3, 3)
sns.countplot(data=df, x='Type_of_order', palette='pastel')
plt.title('Distribution of Type of Order')

# Bar plot for Type_of_vehicle
plt.subplot(2, 3, 4)
sns.countplot(data=df, x='Type_of_vehicle', palette='pastel')
plt.title('Distribution of Type of Vehicle')

# Bar plot for Festival
plt.subplot(2, 3, 5)
sns.countplot(data=df, x='Festival', palette='pastel')
plt.title('Distribution of Festival')

# Bar plot for City
plt.subplot(2, 3, 6)
sns.countplot(data=df, x='City', palette='pastel')
plt.title('Distribution of City')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Weather_conditions', y='Time_taken (min)', palette='pastel')
plt.title('Box plot: Weather_conditions vs Time_taken (min)')
plt.xlabel('Weather_conditions')
plt.ylabel('Time_taken (min)')
plt.show()


```

```python
# Perform ANOVA test
import scipy.stats as stats
anova_weather = stats.f_oneway(*[group['Time_taken (min)'] for name, group in df.groupby('Weather_conditions') if len(group['Time_taken (min)']) > 1])
print("ANOVA p-value for Weather_conditions:", anova_weather.pvalue)
```
